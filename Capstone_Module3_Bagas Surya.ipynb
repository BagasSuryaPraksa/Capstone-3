>> The model will be evaluated using:

> ### Evaluation Metrics
> Kami akan mengevaluasi model menggunakan R^2, RMSE, MAE, dan MAPE. Semakin kecil nilai metrik-metrik ini, semakin akurat model dalam memprediksi nilai Target berdasarkan fitur yang digunakan.

> #### R^2 (R-squared)
> R^2 (R-squared) adalah metrik evaluasi yang mengukur seberapa baik model regresi cocok dengan data. Nilai R^2 berkisar dari 0 hingga 1, di mana 1 menunjukkan bahwa model sepenuhnya cocok dengan data. Semakin tinggi nilai R^2, semakin baik model dapat menjelaskan variasi dalam data.

> #### RMSE (Root Mean Square Error)
> RMSE mengukur sejauh mana perbedaan antara nilai prediksi dari model dan nilai sebenarnya dari data. Metriks ini mengukur deviasi rata-rata antara nilai prediksi dan nilai sebenarnya dengan mengkuadratkan selisihnya, menjumlahkan, dan kemudian mengambil akar kuadrat dari rata-rata tersebut.
> Rumus RMSE adalah sebagai berikut:
> $$RMSE =   \sqrt{ \frac{ \sum\nolimits_{i=1}^{n} ( Y_i - \widehat{Y_i})^2 }{n} } $$
> $$Root Mean Squared Error$$
> - ∑ adalah sigma (penjumlahan dari suatu urutan atau fungsi)<br>
> - ŷi adalah nilai prediksi data ke i<br>
> - yi adalah nilai aktual data ke i<br>
> - n adalah jumlah data<br>
    
> #### MAE (Mean Absolute Error)
> MAE mengukur kesalahan rata-rata absolut antara prediksi dan nilai sebenarnya. Metrik ini tidak sensitif terhadap outlier dan cocok untuk prediksi nilai CLV karena memberikan bobot yang seragam untuk semua perbedaan.
> Rumus MAE adalah sebagai berikut:
> $$MAE =  \frac{\sum\nolimits_{i=1}^{n} |Y_i - \widehat{Y_i}|}{n}$$
> $$Mean Absolute Error$$
> - ∑ adalah sigma (penjumlahan dari suatu urutan atau fungsi)<br>
> - ŷi adalah nilai prediksi data ke i<br>
> - yi adalah nilai aktual data ke i<br>
> - n adalah jumlah data<br>

> #### MAPE (Mean Absolute Percentage Error)
> MAPE mengukur kesalahan dalam bentuk persentase dari nilai sebenarnya, memberikan gambaran tentang akurasi model dalam persentase.
> Rumus MAPE adalah sebagai berikut:
> $$MAPE =  \frac{1}{n}   \sum_{i=1}^{n} | \frac{Y_i - \widehat{Y_i}}{Y_i}|$$
> $$Mean Absolute Percentage Error$$
> - ∑ adalah sigma (penjumlahan dari suatu urutan atau fungsi)<br>
> - ŷi adalah nilai prediksi data ke i<br>
> - yi adalah nilai aktual data ke i<br>
> - n adalah jumlah data<br>

```python
# Seperating features by its data types
numeric_features = ['Size(sqf)', 'N_FacilitiesNearBy(ETC)', 'N_FacilitiesNearBy(PublicOffice)', 'N_SchoolNearBy(University)', 
                    'N_Parkinglot(Basement)', 'N_FacilitiesInApt']
categorical_features = ['HallwayType', 'SubwayStation']
ordinal_features = ['TimeToSubway']

# Mapping for ordinal encoders (TimeToSubway)
ordinal_map_subway = [{'col':'TimeToSubway',
                'mapping':{'no_bus_stop_nearby':1, 
                           '15min~20min':2, 
                           '10min~15min':3, 
                           '5min~10min':4,
                           '0-5min':5}}]                           
```

**MULTICORELATION ANALYSIS**

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_no_outliers.drop(columns=['SalePrice']).select_dtypes(include='number').corr(method='spearman'),annot=True)
plt.show()
```

**Pipeline, encoding dan scaling**

```python
# Preprocessor for encoding dan scaling
preprocessor = ColumnTransformer(
    transformers=[
        ('num', RobustScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('ord', ce_OrdinalEncoder(mapping=ordinal_map_subway), ordinal_features)
    ],
    remainder='passthrough'
)

# Pipeline for feature engineering
FE_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
])

FE_pipeline
```

```python
#make pipeline dataframe
df_transformed = pd.DataFrame(FE_pipeline.fit_transform(df_no_outliers))
df_transformed
```

```python
#rename columns after transformation
FE_pipeline.named_steps['preprocessor'].transformers_
```

```python
# Ordinal features Subway
FE_pipeline.named_steps['preprocessor'].transformers_[2][1].get_feature_names_out(ordinal_features)
```

```python
# Remaining features
remaining_columns = df_no_outliers.drop(numeric_features + categorical_features + ordinal_features, axis=1).columns
remaining_columns
```

```python
# Combine all features
all_features = list(FE_pipeline.named_steps['preprocessor'].transformers_[0][2]) + \
                list(FE_pipeline.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(categorical_features)) + \
                list(FE_pipeline.named_steps['preprocessor'].transformers_[2][1].get_feature_names_out(ordinal_features)) + \
                list(remaining_columns)
                
df_transformed.columns = all_features
df_transformed
```

Feature Engineering adalah proses membuat fitur-fitur baru atau mengubah fitur-fitur yang ada dengan tujuan meningkatkan kinerja model. Proses Feature Engineering dalam proyek ini menggunakan pustaka Pipeline dan deskripsinya adalah sebagai berikut:

- **RobustScaler**: Robust Scaler adalah teknik penskalaan fitur berdasarkan statistik yang kurang sensitif terhadap pencilan (median dan rentang interkuartil) dan umumnya digunakan untuk fitur-fitur numerik. Penskalaan ini membantu model agar tidak terpengaruh oleh nilai ekstrem yang mungkin mempengaruhi/menyesatkan model.

- **OneHotEncoder**: OneHotEncoder digunakan untuk mengonversi variabel kategorikal menjadi serangkaian variabel biner (numerik 1 atau 0) atau juga dapat disebut variabel dummy. Tujuannya adalah untuk memudahkan pemrosesan informasi kategorikal tanpa mengasumsikan urutan antara kategori-kategori (nominal).

- **OrdinalEncoder**: OrdinalEncoder menerjemahkan nilai-nilai seperti 'TimeToSubway' ke dalam skala numerik ordinal sesuai dengan pemetaan yang telah ditentukan sebelumnya. Fitur ordinal adalah fitur kategori yang memiliki urutan atau tingkatan.

- **Passthrough**: passthorugh digunakan untuk mengabaikan fitur-fitur yang tidak ingin diubah (dalam pra-pemrosesan) karena fitur-fitur ini sudah dianggap sesuai untuk model yang akan dijalankan. Pada model ini, kami membiarkan kolom YearBuilt seperti adanya.

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error, mean_squared_error
```

```python
# define function for model evaluation

def evaluate_model(model, X_train, X_test, y_train, y_test, name):
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions and evaluate
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)
    
    mae_train = mean_absolute_error(y_train, train_preds)
    mae_test = mean_absolute_error(y_test, test_preds)
    
    r2_train = r2_score(y_train, train_preds)
    r2_test = r2_score(y_test, test_preds)
    
    mape_train = mean_absolute_percentage_error(y_train, train_preds)
    mape_test = mean_absolute_percentage_error(y_test, test_preds)
    
    rmse_train = mean_squared_error(y_train, train_preds, squared=False)
    rmse_test = mean_squared_error(y_test, test_preds, squared=False)
    
    # Save the results and create a dataframe
    data = {
        'Training R^2': r2_train,
        'Test R^2': r2_test,
        'Training MAE': mae_train,
        'Test MAE': mae_test,
        'Training MAPE': mape_train,
        'Test MAPE': mape_test,
        'Training RMSE': rmse_train,
        'Test RMSE': rmse_test
    }
    
    df_results = pd.DataFrame(data, index=[name])
    
    return df_results
```

```python
linear = evaluate_model(LinearRegression(), X_train, X_test, y_train, y_test, 'Linear Regression')
lasso = evaluate_model(Lasso(random_state=99), X_train, X_test, y_train, y_test, 'Lasso')
ridge = evaluate_model(Ridge(random_state=99), X_train, X_test, y_train, y_test, 'Ridge')
knn = evaluate_model(KNeighborsRegressor(), X_train, X_test, y_train, y_test, 'KNN')
tree = evaluate_model(DecisionTreeRegressor(random_state=99), X_train, X_test, y_train, y_test, 'Decision Tree')
forest = evaluate_model(RandomForestRegressor(random_state=99), X_train, X_test, y_train, y_test, 'Random Forest')
svr = evaluate_model(SVR(), X_train, X_test, y_train, y_test, 'SVR')
```

```python
df_eval = pd.concat([linear, lasso, ridge, knn, tree, forest, svr])
df_eval.sort_values(by='Test R^2', ascending=False)
```

### Definition (How it Work)


### Ensemble Method

- Pembelajaran ensemble adalah teknik yang menggabungkan prediksi dari beberapa algoritma machine learning untuk menghasilkan prediksi yang lebih akurat daripada model individu mana pun. Dengan kata lain, model ensemble adalah model yang terdiri dari banyak model. Ada banyak teknik ensemble seperti Stacking, Bagging, Boosting, dll.

### Bagging (Bootstrap Aggregating)

Bootstrapping adalah teknik sampling di mana kita membuat subset observasi dari dataset asli, teknik ini juga disebut Bagging. Dalam teknik ini, hasil yang umum diperoleh dengan menggabungkan hasil dari berbagai model prediktif. Ini menciptakan subset pelatihan yang berbeda dari data pelatihan sampel dengan penggantian & output final didasarkan pada voting mayoritas. Sebagai contoh, **Random Forest**

. Berikut langkah-langkah yang terlibat dalam Bagging:

  1. **Selection of Subset**: Bagging dimulai dengan memilih sampel acak, atau subset, dari seluruh dataset.
  2. **Bootstrap Sampling**: Setiap model kemudian dibuat dari sampel-sampel ini, yang disebut Bootstrap Samples, yang diambil dari data asli dengan penggantian. Proses ini dikenal sebagai pengambilan sampel baris.
  3. **Independent Model Training**: Setiap model dilatih secara independen pada Bootstrap Sample yang sesuai. Proses pelatihan ini menghasilkan hasil untuk setiap model.
  4. **Majority Voting**: Output final ditentukan dengan menggabungkan hasil dari semua model melalui voting mayoritas. Hasil yang paling sering diprediksi di antara model-model dipilih.
  5. **Aggregation**: Langkah ini, yang melibatkan menggabungkan semua hasil dan menghasilkan output final berdasarkan voting mayoritas, dikenal sebagai agregasi.
    
### Steps Involved in Random Forest Algorithm
- **Step 1**: Dalam model Random forest, subset dari titik data dan subset dari fitur dipilih untuk membangun setiap pohon keputusan. Secara sederhana, n-record acak dan m-fitur diambil dari kumpulan data memiliki k-record.
- **Step 2**: Pohon keputusan individual dibangun untuk setiap sampel.
- **Step 3**: Setiap pohon keputusan akan menghasilkan output.
- **Step 4**: Output akhir dipertimbangkan berdasarkan Voting Mayoritas atau Rata-rata untuk Klasifikasi dan Regresi, secara berurutan.

Dalam istilah sederhana, **Random Forest adalah kumpulan hasil prediksi dari beberapa Pohon Keputusan**. Jadi dapat dikatakan bahwa hasil prediksi dari Random Forest akan lebih kokoh dan akurat jika dibandingkan dengan hasil prediksi dari satu Pohon Keputusan.

## 7. Hyper Parameter Tuning

- RandomizedSearchCV secara acak melalui serangkaian hiperparameter dan menghitung skor serta memberikan set hiperparameter terbaik yang memberikan skor terbaik sebagai output. Jadi berikut adalah resep Bagaimana kita dapat menemukan parameter optimal menggunakan RandomizedSearchCV untuk Regresi.

- Perhatikan bahwa RandomizedSearchCV() memerlukan argumen tambahan n_iter, yang menentukan berapa banyak sel acak yang harus dipilih. Ini menentukan berapa kali validasi silang k akan dilakukan.

- Validasi silang adalah prosedur resampling yang digunakan untuk mengevaluasi model machine learning pada sampel data terbatas. Jika Anda memiliki model machine learning dan beberapa data, Anda ingin tahu apakah model Anda dapat cocok. Anda dapat membagi data Anda menjadi set pelatihan dan uji.



### 7.1 Random Forest Tuning

```python
# Hyperparameter tuning Random Forest using RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200]
# Number of features to consider at every split
max_features = ['log2', 'sqrt', None]
# Maximum number of levels in tree
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10, None]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Max samples for bootstrapping
max_samples = [0.5, 0.7, 0.8, 1.0]

# Create the random grid
random_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'max_samples': max_samples
}

# Use the random grid to search for best hyperparameters

# First create the base model to tune
rf = RandomForestRegressor(random_state=99)

# Random search of parameters, using 3 fold cross validation,
# search across 200 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator=rf,
                               param_distributions=random_grid,
                               n_iter=200,
                               cv=3,
                               random_state=99,
                               n_jobs=-1)
```

```python
rf_random.fit(X_train, y_train)
```

```python
results_rf_random = pd.DataFrame(rf_random.cv_results_)
results_rf_random.sort_values(by='rank_test_score').head(2)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
rf_random.best_estimator_
```

```python
# Evaluate the Random Forest model with the best hyperparameters
y_pred_rf_random = rf_random.best_estimator_.predict(X_test)

r2_rf_random = r2_score(y_test, y_pred_rf_random)
mae_rf_random = mean_absolute_error(y_test, y_pred_rf_random)
mape_rf_random = mean_absolute_percentage_error(y_test, y_pred_rf_random)
rmse_rf_random = mean_squared_error(y_test, y_pred_rf_random, squared=False)

df_forest_tune = pd.DataFrame({'Random Forest RandomizedSearchCV': [r2_rf_random, mae_rf_random, mape_rf_random, rmse_rf_random]}, index=['R^2', 'MAE', 'MAPE', 'RMSE'])
df_forest_tune
```

Berikut adalah metrik yang digunakan untuk mengevaluasi kinerja model Regresi Random Forest dalam memprediksi harga apartemen dalam Won Korea (₩). Berikut adalah cara menginterpretasikannya:

3. MAPE (Mean Absolute Percentage Error): Ini adalah 0,193297, atau 19,33%. Ini berarti secara rata-rata, prediksi model Anda sekitar 19,33% dari harga apartemen sebenarnya. Ini memberikan ukuran relatif dari kesalahan yang dibuat oleh model.

4. RMSE (Root Mean Square Error): Ini adalah 42.977,042072, yang merupakan ukuran dari perbedaan antara nilai-nilai yang diprediksi oleh model dan nilai-nilai yang sebenarnya diamati. Ini lebih sensitif terhadap kesalahan besar (Outlier) daripada MAE karena mengkuadratkan perbedaan sebelum rata-rata. RMSE yang lebih besar dari MAE menunjukkan bahwa model Anda kadang-kadang membuat kesalahan besar.

MAE dan RMSE dapat digunakan bersama untuk mendiagnosis variasi dalam kesalahan dalam satu set ramalan. RMSE akan selalu lebih besar atau sama dengan MAE; semakin besar perbedaan di antara keduanya, semakin besar variasi dalam kesalahan individual dalam sampel. Jika RMSE = MAE, maka semua kesalahan memiliki magnitudo yang sama (₩42,977.04 - ₩34,966.13 = ₩8,010.91 => Perbedaan antara MAE dan RMSE tidak terlalu besar, artinya dapat dikatakan bahwa variasi kesalahan dalam hasil prediksi tidak terlalu besar.)
Secara keseluruhan, metrik-metrik ini menunjukkan bahwa model ini melakukan pekerjaan yang cukup baik dalam memprediksi harga apartemen dalam Won Korea, tetapi masih ada ruang untuk perbaikan, terutama dalam mengurangi besarnya kesalahan.

Namun, penting untuk diakui bahwa model ini memiliki batasan. Prediksi terbatas oleh rentang dan kategori fitur dalam dataset pelatihan. Terutama, akurasi model dibatasi oleh nilai minimum dan maksimum fitur seperti luas apartemen, dan hanya dapat memprediksi dalam batasan kategori yang dipelajari (misalnya, tipe lorong, stasiun kereta bawah tanah).
